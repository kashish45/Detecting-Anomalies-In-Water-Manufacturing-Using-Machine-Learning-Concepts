```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns',None)
```

```python
train_data=pd.read_csv('Train.csv')
```

```python
train_data.shape
```

```python
train_data.head()
```

```python
train_data['Class'].value_counts()
```

```python
X=train_data.iloc[:,:-1]
y=train_data.iloc[:,-1]
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
#Handling Imbalance Dataset using CROSS VALIDATION AND K FOLD
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import KFold
import numpy as np
from sklearn.model_selection import GridSearchCV
```

```python
log_class=LogisticRegression()
grid={'C':10.0 **np.arange(-2,3),'penalty':['l1','l2']}
cv=KFold(n_splits=5,shuffle=False,random_state=None)
```

```python
clf=GridSearchCV(log_class,grid,cv=cv,n_jobs=-1,scoring='f1_macro')
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred))
print("\n")
print("accuracy Score:")
print(accuracy_score(y_test,y_pred))
print("\n")
print(classification_report(y_test,y_pred))
```

```python
class_weight=dict({0:1,1:100}) #GIVING MORE IMPORTANCE TO 1 
```

```python
from sklearn.ensemble import RandomForestClassifier
clasf=RandomForestClassifier(class_weight=class_weight)
clasf.fit(X_train,y_train)
clasf.score(X_train,y_train)
```

```python
y_predd=clasf.predict(X_test)
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_predd))
print("\n")
print("accuracy Score:")
print(accuracy_score(y_test,y_predd))
print("\n")
print(classification_report(y_test,y_predd))
```

```python
#UnderSampling
```

```python
from collections import Counter
from imblearn.under_sampling import NearMiss
ns=NearMiss(0.6) #Reducing Points by 80%
X_train_ns,y_train_ns=ns.fit_sample(X_train,y_train)
print("the number of classes before fit {}".format(Counter(y_train)))
print("the number of classes after fit {}".format(Counter(y_train_ns)))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train_ns,y_train_ns)
```

```python
classifier.score(X_train_ns,y_train_ns)
```

```python
y_prd=classifier.predict(X_test)
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_prd))
print("\n")
print("accuracy Score:")
print(accuracy_score(y_test,y_prd))
print("\n")
print(classification_report(y_test,y_prd))
```

```python
# #FiLTER METHODS FOR FEATURE SELECTION
# Features selected using filter methods can be used as an input to any machine learning models.

# Univariate -> Fisher Score, Mutual Information Gain, Variance etc
# Multi-variate -> Pearson Correlation
```

```python
# #The univariate filter methods are the type of methods where individual features are ranked according to specific criteria. 
# The top N features are then selected. Different types of ranking criteria are used for univariate filtermethods, for example 
# fisher score, mutual information, and variance of the feature.
```

```python
# Multivariate filter methods are capable of removing redundant features from the 
# data since they take the mutual relationship between the features into account.
```

```python
#Univariate MEthods
```

```python
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import VarianceThreshold
```

```python
constant_filter = VarianceThreshold(threshold=0)
constant_filter.fit(X_train)
```

```python
constant_filter.get_support().sum() #No of features left after Removing COnstant Features
```

```python
X_train_filter = constant_filter.transform(X_train)
X_test_filter = constant_filter.transform(X_test)
```

```python
X_train_filter.shape, X_test_filter.shape, X_train.shape
```

```python
quasi_constant_filter = VarianceThreshold(threshold=0.01)
quasi_constant_filter.fit(X_train_filter)
VarianceThreshold(threshold=0.01)
```

```python
quasi_constant_filter.get_support().sum()
```

```python
X_train_quasi_filter = quasi_constant_filter.transform(X_train_filter)
X_test_quasi_filter = quasi_constant_filter.transform(X_test_filter)
X_train_quasi_filter.shape, X_test_quasi_filter.shape
```

```python
X_train_T = X_train_quasi_filter.T
X_test_T = X_test_quasi_filter.T
type(X_train_T)
```

```python
X_train_T = pd.DataFrame(X_train_T)
X_test_T = pd.DataFrame(X_test_T)
```

```python
X_train_T.duplicated().sum()
```

```python
duplicated_features = X_train_T.duplicated()
duplicated_features
```

```python
features_to_keep = [not index for index in duplicated_features]
```

```python
X_train_unique = X_train_T[features_to_keep].T
X_test_unique = X_test_T[features_to_keep].T
```

```python
X_train_unique.shape, X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
clf2=RandomForestClassifier(class_weight=class_weight)
clf2.fit(X_train_unique,y_train)
clf2.score(X_train_unique,y_train)
```

```python
y_prd=clf2.predict(X_test_unique)
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_prd))
print("\n")
print("accuracy Score:")
print(accuracy_score(y_test,y_prd))
print("\n")
print(classification_report(y_test,y_prd))
```

```python
os=RandomOverSampler(0.95)
X_train_os1,y_train_os1=os.fit_sample(X_train_unique,y_train)
print("the number of classes before fit {}".format(Counter(y_train)))
print("the number of classes after fit {}".format(Counter(y_train_os1)))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf3=RandomForestClassifier()
clf3.fit(X_train_os1,y_train_os1)
clf3.score(X_train_os1,y_train_os1)
```

```python
y_prd=clf3.predict(X_test_unique)
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_prd))
print("\n")
print("accuracy Score:")
print(accuracy_score(y_test,y_prd))
print("\n")
print(classification_report(y_test,y_prd))
```
